```python

import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as ec
from selenium.webdriver.common.by import By
import matplotlib.pyplot as plt

from matplotlib.font_manager import FontProperties
fontP = FontProperties()
fontP.set_size('small') 

# Configuración directa del driver
options = webdriver.ChromeOptions()
prefs = {'download.default_directory': '../Series/Ficheros'}
options.add_experimental_option('prefs', prefs)
options.add_experimental_option('excludeSwitches', ['enable-logging'])

# Instala o actualiza el driver automáticamente
driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=options
)

```

```python
def tablas_df(texto):
    tabla_html  = texto.replace('\t', ';')
    tabla_html  = tabla_html.replace('¹', '')

    # Dividir el texto en líneas
    lineas = tabla_html.split('\n')

    # Dividir cada línea en elementos usando punto y coma como separador
    lista_resultante = [linea.split(';') for linea in lineas]


    # Crear un diccionario a partir de la lista
    diccionario_resultante = {}
    for elemento in lista_resultante:
        clave = elemento[0]
        valores = elemento[1:]
        diccionario_resultante[clave] = valores

    df = pd.DataFrame(diccionario_resultante).T


    # Nombres de las columnas
    nombres_columnas = ['Andalucía.dato', 'España.dato', 'Andalucía.Tasa', 'España.tasa', 'Período', '']

    # Asignar nombres a las columnas
    df.columns = nombres_columnas

    return df
```

```python
from io import StringIO
url=  'https://www.juntadeandalucia.es/institutodeestadisticaycartografia/ultimosDatos/index.html'

# Navegar a la página web
driver.get(url)

# Esperar a que la página cargue completamente (puedes ajustar el tiempo según sea necesario)
driver.implicitly_wait(10)

#OFERTA

# Encontrar el elemento que contiene la tabla
tabla_element = driver.find_element(By.XPATH, '//*[@id="seccion2"]/div[2]/table/tbody')
# Obtener el HTML de la tabla
tabla = tabla_element.get_attribute('innerText')
df_oferta= tablas_df(tabla)


#DEMANDA


# Encontrar el elemento que contiene la tabla
tabla_element = driver.find_element(By.XPATH, '//*[@id="seccion2"]/div[3]/table/tbody')
# Obtener el HTML de la tabla
tabla = tabla_element.get_attribute('innerText')
df_demanda= tablas_df(tabla)

#FINANCIEROS


# Encontrar el elemento que contiene la tabla
tabla_element = driver.find_element(By.XPATH, '//*[@id="seccion2"]/div[4]/table/tbody')
# Obtener el HTML de la tabla
tabla = tabla_element.get_attribute('innerText')
df_financieros= tablas_df(tabla)


#PRECIOS


# Encontrar el elemento que contiene la tabla
tabla_element = driver.find_element(By.XPATH, '//*[@id="seccion2"]/div[5]/table/tbody')
# Obtener el HTML de la tabla
tabla = tabla_element.get_attribute('innerText')
df_precios= tablas_df(tabla)


#EMPRESAS


# Encontrar el elemento que contiene la tabla
tabla_element = driver.find_element(By.XPATH, '//*[@id="seccion2"]/div[6]/table/tbody')
# Obtener el HTML de la tabla
tabla = tabla_element.get_attribute('innerText')
df_empresas= tablas_df(tabla)


#COMERCIO EXTERIOR


# Encontrar el elemento que contiene la tabla
tabla_element = driver.find_element(By.XPATH, '//*[@id="seccion2"]/div[7]/table/tbody')
# Obtener el HTML de la tabla
tabla = tabla_element.get_attribute('innerText')
df_exterior= tablas_df(tabla)


df = pd.concat([df_oferta, df_demanda, df_financieros, df_precios, df_empresas, df_exterior])

df.to_excel('../Series/Tabla/outputII.xlsx')
```
